```python
%matplotlib inline

import numpy as np
import pandas as pd
from pandas_datareader import data
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
from_date = '2014-06-30'
to_date = '2015-06-30'

asset1 = data.DataReader('HSC','yahoo',from_date,to_date)['Close'].pct_change()[1:]
asset2 = data.DataReader('MSFT','yahoo',from_date,to_date)['Close'].pct_change()[1:]
bench = data.DataReader('SPY','yahoo',from_date,to_date)['Close'].pct_change()[1:]
tresury_ret = data.DataReader('BIL','yahoo',from_date,to_date)['Close'].pct_change()[1:]
constant = pd.Series(np.ones(len(asset1.index)),index=asset1.index)
```

```python
dt = pd.DataFrame({'R1': asset1, 'R2' : asset2, 'SPY' : bench, 'RF' : tresury_ret, 'Constant' : constant})
dt.dropna()
```

```python
reg1 = LinearRegression()
reg1.fit(X=np.array( [ dt['SPY'].tolist() , dt['RF'].tolist() ] ).T,y=np.array([dt['R1'].tolist()]).T)
print reg1.coef_
print reg1.intercept_

reg2 = LinearRegression()
reg2.fit(X=np.array( [ dt['SPY'].tolist() , dt['RF'].tolist() ] ).T,y=np.array([dt['R2'].tolist()]).T)
print reg2.coef_
print reg2.intercept_
```

```python
model = pd.stats.ols.MovingOLS(y = dt['R2'], x=dt[['SPY', 'RF']], 
                             window_type='rolling', window=100)
rolling_parameter_estimates = model.beta
rolling_parameter_estimates.plot()

plt.title('Asset1 Computed Betas')
plt.legend(['Market Beta', 'Risk Free Beta', 'Intercept', 'Market Beta Static', 'Risk Free Beta Static', 'Intercept Static'])
```
